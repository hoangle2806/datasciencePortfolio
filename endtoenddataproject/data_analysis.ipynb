```python
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('processeddata/creditcard.txt')
df.head()
```

```python
y = df['Class']
features = [i for i in df.columns if i != 'Class']
x = df[features]
x.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.4, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics
clf = RandomForestClassifier(n_estimators = 10)

clf = clf.fit(x_train,y_train)
```

```python
predictions = clf.predict(x_test)
print "accuracy score: {}".format(metrics.accuracy_score(y_test,predictions))
```

```python
from sklearn.preprocessing import StandardScaler
df['normalizedAmount'] = StandardScaler().fit_transform(df['Amount'].reshape(-1,1))
df = df.drop(['Amount'], axis =1)
df.head()
```

```python
y = df['Class']
features = [i for i in df.columns if i != 'Class']
x = df[features]
x.head()
```

```python
import numpy as np
fraud_class = len(df[df['Class']==1])
fraud_indices = np.array(df[df['Class'] == 1].index)

#indices for the normal classes (non-fraud)
normal_case_indices = df[df['Class'] == 0].index

#randomly pick x numbers of fraud
random_normal = np.random.choice(normal_case_indices, fraud_class, replace = False)
random_normal = np.array(random_normal)

#combine 2 indices
subsample_indices = np.concatenate([fraud_indices,random_normal])
#subsample
subsample_data = df.iloc[subsample_indices,:]

x_sub = subsample_data.ix[:,subsample_data.columns != 'Class']
y_sub = subsample_data.ix[:,subsample_data.columns == 'Class']

#check ratio and sample of data

print "Percent of normal transactions: %f" % (len(x_sub)*100.0/len(df))
print "Percent of fraud transactions: %f" % (len(y_sub)*100.0/len(df))
print "number of transactions resample: %s" % len(subsample_data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sub,y_sub, test_size = 0.4, random_state = 0)
```

```python
clf = RandomForestClassifier(n_estimators = 10)

clf = clf.fit(x_train,y_train)
predictions = clf.predict(x_test)
print "accuracy score: {}".format(metrics.accuracy_score(y_test,predictions))
```
